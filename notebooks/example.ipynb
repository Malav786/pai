### DATA LOADING

```python
from sklearn.datasets import fetch_lfw_people
```

```python
lfw_people = fetch_lfw_people(min_faces_per_person=70, resize=0.4)
for name in lfw_people.target_names:
    print(name)
```

```python
print(lfw_people.images.shape)   # (n_samples, height, width)
print(lfw_people.target.shape)   # labels
print(lfw_people.target_names)   # person names
```

```python
import matplotlib.pyplot as plt

# load first 5 images
for i in range(5):
    plt.imshow(lfw_people.images[i], cmap='gray')
    plt.title(lfw_people.target_names[lfw_people.target[i]])
    plt.axis('off')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

person_name = "George W Bush"
person_id = np.where(lfw_people.target_names == person_name)[0][0]

# get indices of images belonging to this person
indices = np.where(lfw_people.target == person_id)[0]

# show first 3 images
for i in indices[:3]:
    plt.imshow(lfw_people.images[i], cmap='gray')
    plt.title(person_name)
    plt.axis('off')
    plt.show()
```

```python
import numpy as np

for i, name in enumerate(lfw_people.target_names):
    count = np.sum(lfw_people.target == i)
    print(f"{name}: {count} images")
```
